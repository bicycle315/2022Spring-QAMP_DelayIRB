```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
belem_x=[0.000186,0.0002449]
belem_d=[0.000201, 0.000283]
belem_cle=[0.000164, 0.000129]

toronto_x=[0.0004924, 0.0002647]
toronto_d=[0.000128,0.000212]
toronto_cle=[0.000185,0.000163]

guadalupe_x=[0.000372]
guadalupe_d=[0.000397]
guadalupe_cle=[0.000266]

nairobi_x=[0.0003205, 0.0002827, 0.0002901,0.0004450]
nairobi_d=[0.0003446,0.0003034, 0.0003078,0.0008927]
nairobi_cle=[0.0002759,0.0001469,0.0001712,0.0002303]

quito_x=[0.0003255]
quito_d=[0.0003067]
quito_cle=[0.0001372]
```

```python
fig = plt.figure(figsize=(12,6))
ax2 = fig.add_subplot(1,2,1)
ax3 = fig.add_subplot(1,2,2)


ax2.scatter(guadalupe_x, guadalupe_d, c='orange', label='guadalupe_q0')
ax2.scatter(belem_x,belem_d, c='yellow', label='belem_q0,3')
ax2.scatter(nairobi_x,nairobi_d, c='green', label='nairobi_0,1,4,5')
ax2.scatter(toronto_x,toronto_d, c='blue',label='toronto_q1,8')

ax2.set_ylabel('Delay', c='black')
ax2.set_xlabel('X', c='black')
ax2.text(0.0004, 0.0006, 'Delay>X')
ax2.text(0.0007, 0.0006, 'Delay<X')
ax2.set_ylim(0,0.001)
ax2.set_xlim(0,0.001)
ax2.plot([0,0.01],[0,0.01], color='black')
ax2.legend(loc='lower right')
ax2.set_title('X,Delay Correlation')

ax3.scatter(guadalupe_x, guadalupe_cle, c='orange', label='guadalupe_q0')
ax3.scatter(belem_x,belem_cle, c='yellow', label='belem_q0,3')
ax3.scatter(nairobi_x,nairobi_cle, c='green', label='nairobi_0,1,4,5')
ax3.scatter(toronto_x,toronto_cle, c='blue',label='toronto_q1,8')



ax3.set_ylim(0,0.001)
ax3.set_xlim(0,0.001)
ax3.set_ylabel('CLE', c='black')
ax3.plot([0,0.01],[0,0.01], color='black')
ax3.set_xlabel('X', c='black')
ax3.text(0.0008, 0.0003, 'CLE<X')
ax3.legend(loc='upper left')
ax3.set_title('X,CLE Correlation')

plt.show()
```
